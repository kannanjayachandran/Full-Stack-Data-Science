```python
import matplotlib.pyplot as plt
import numpy as np
```

## <font color="Orange"> Basics</font>

```python
arr1 = [1, 2, 4, 5, 4, 3, 2, 1]


plt.plot(arr1)
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("Plot 1")

# plt.show()    # we can avoid this in jupyter notebooks

# Here since we didn't mention the `x` values, indices of the `y` values are taken as x-values
```

```python
arr2 = [x for x in range(1, 11)]
arr3 = [x**2 for x in range(1, 11)]


plt.plot(arr2, arr3)
plt.ylabel("Numbers")
plt.xlabel("Squares")
plt.title("Plot 2")
plt.grid() # grid for graph
plt.show()

# points instead of line
plt.plot(arr2, arr3, 'ro')
plt.ylabel("Numbers")
plt.xlabel("Squares")
plt.title("Plot 3")
plt.grid()
```

```python
nums = np.arange(1, 10, 0.5)

# multiple plots in one graph
plt.plot(nums, nums**2, 'b--', label='square')
plt.plot(nums, nums**0.5, 'rs', label='sq_root')
plt.plot(nums, nums**3, 'g^', label='cube')

plt.xlabel('Numbers')
plt.ylabel('root-square-cube')
plt.title('Plot 4')
plt.legend()    # show the legend inside the graph
plt.grid()
```

```python
# controlling the line
x = [x for x in range(1, 11)]
y = [x**2 for x in x]

plt.plot(x, y, 'r--', linewidth=7.0)
plt.plot(x, y, 'bo', markersize=20)
plt.title('Plot 5')
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
x2 = [3, 4, 5, 6, 7]
y2 = [4, 5, 6, 7, 8]

lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines[0], color='g', linewidth=10.0)
plt.setp(lines[1], color='r', linewidth=5.0)
plt.title('Plot 6')
```

```python
# subplots

def foo(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)       # We need one figure 
plt.subplot(211)    # Subplot(rows, cols, figure_number) : figure_number can range from 1 to rows*cols
plt.plot(t1, foo(t1), 'g--')
plt.grid()

plt.subplot(212)
plt.plot(t2, foo(t2))

# grouping plots
plt.figure(2)
plt.subplot(331)
plt.plot(t1, foo(t1), 'g--')
plt.grid()

plt.figure(2)
plt.subplot(332)
plt.plot(t1, foo(t1), 'r--')
plt.grid()

plt.figure(2)
plt.subplot(333)
plt.plot(t1, foo(t1), 'b--')
plt.grid()

plt.figure(2)
plt.subplot(334)
plt.plot(t1, foo(t1), 'ro')
plt.grid()

plt.figure(2)
plt.subplot(336)
plt.plot(t1, foo(t1), 'bo')
plt.grid()

plt.figure(2)
plt.subplot(337)
plt.plot(t1, foo(t1), 'go')
plt.grid()

plt.figure(2)
plt.subplot(338)
plt.plot(t1, foo(t1))
plt.grid()

plt.figure(2)
plt.subplot(339)
plt.plot(t1, foo(t1), 'g')
plt.grid()
```

## <font color="Orange"> Scatter plot </font>

```python
data_x = np.random.random(25) * 100
data_y = np.random.random(25) * 100

plt.scatter(data_x, data_y)
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('d', 'a', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
x_data = np.random.random(2500) * 100
y_data = np.random.random(2500) * 100

plt.scatter(x_data, y_data, c="orange", marker="*", s=120, alpha=0.3)
```

## <font color="Orange"> Bar plot </font>

```python
data_x_axis = ['one', 'two', 'three', 'four', 'five']
data_y_axis = [1, 2, 3, 4, 5]

plt.bar(data_x_axis, data_y_axis)
plt.xlabel('Numbers')
plt.ylabel('Numbers')
plt.title('Plot 10')
plt.show()
```

## <font color="Orange"> Histogram </font>

```python
mean = 30
std_deviation = 1.5
samples = 1000
salary_per_hr = np.random.normal(mean, std_deviation, samples)

plt.xlabel('Hours')
plt.ylabel('Salary')
plt.title('Salary/Hour plot')
plt.hist(salary_per_hr, bins=20, facecolor='g')
plt.show()
```

```python
plt.xlabel('Hours')
plt.ylabel('Salary')
plt.title('Salary/Hour plot 2')
plt.hist(salary_per_hr, bins=20, cumulative=True, color="orange")
plt.show()
```

## <font color="Orange"> 3D Plotting </font>

```python
ax = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title("3D Plot")
```
