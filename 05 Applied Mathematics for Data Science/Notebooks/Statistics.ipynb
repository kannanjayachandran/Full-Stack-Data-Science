```python
import numpy as np
rng = np.random.default_rng(369)
```

## Summary statistics

- Mean
- Median
- Mod
- Variance
- Standard Deviation
- Range

```python
def create_random_data(start: int, end: int, size: int) -> list[int]:
    return rng.integers(start, end, size)


library_dataset = create_random_data(1, 9, 16)
dataset = create_random_data(1, 9, 16)
sample_data = create_random_data(1, 9, 16)
```

```python
# Library Functions
print("Dataset is : ", library_dataset)

# mean
print(f'Mean : {np.mean(library_dataset):.3f}')

# median
print(f'Median : {np.median(library_dataset):.3f}')

# variance
print(f'Variance : {np.var(library_dataset):.3f}')

# standard deviation
print(f'Std. Deviation : {np.std(library_dataset):.3f}')

# co-eff of standard deviation
print(f'Co-eff. of Std. Deviation : {np.std(library_dataset)/np.mean(library_dataset):.5f}')

# Percentage
percentage_lis = (library_dataset / np.sum(library_dataset)) * 100
percentage_lis = np.round(percentage_lis, decimals=2)
print("Percentage array : ", percentage_lis)

# Percentile
print(f"25th Percentile of the array : {np.percentile(library_dataset, 25)}")
```

```python
from collections import Counter
from math import sqrt


print("Dataset is : ", dataset)

# mean
n = len(dataset)
mean = sum(dataset) / n
print(f"Mean : {mean:.2f}")

# median
sorted_dataset = sorted(dataset)
if n % 2 == 0 :
    d1 = sorted_dataset[n//2]
    d2 = sorted_dataset[n // 2 - 1]
    median = (d1 + d2) / 2
else:
    median = sorted_dataset[n//2]
print(f"Median is : {median}")

# Mode
dic = dict(Counter(dataset))
mod = [int(k) for k, v in dic.items() if v == max(list(dic.values()))]

if len(mod) == n:
    print("No mod found")
else:
    print("Mod is/are : ", mod)
```

```python
print("Dataset : ", sample_data)

# variance
var_lis = [(x-mean)**2 for x in sample_data]
variance = sum(var_lis) / n
print(f'Variance : {variance:.3f}')


# standard deviation
std_dev = sqrt(variance)
print(f'Standard deviation : {std_dev:.3f}')

# Range
range_stat = max(sample_data) - min(sample_data)
print("Range : ", range_stat)

# Percentage {values > 5}
count = sum(1 for num in sample_data if num > 5)
percentage = (count / len(sample_data)) * 100
print("Percentage of values greater than 5 : ", percentage)
```
