# Calculus

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

1. Predicting house prices where price increases linearly with square footage:
$$\text{Price} = 50000 + 200 \times \text{SquareFeet}$$

```python
# Linear function: f(x) = 2x + 3
def linear_function(x, m=2, b=3):
    """
    Linear function implementation
    Args:
        x: input values
        m: slope
        b: y-intercept
    """
    return m * x + b

x = np.linspace(-10, 10, 100)
y = linear_function(x)

plt.plot(x, y, label='f(x) = 2x + 3')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.legend()
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

2. Modeling the trajectory of gradient descent in a simple quadratic loss function.

```python
# Quadratic function: f(x) = x^2 - 4x + 3
def quadratic_function(x, a=1, b=-4, c=3):
    """
    Quadratic function (parabola)
    Vertex at x = -b/(2a)
    """
    return a * x**2 + b * x + c

x = np.linspace(-1, 5, 100)
y = quadratic_function(x)

# Find vertex
vertex_x = -(-4) / (2 * 1)
vertex_y = quadratic_function(vertex_x)

plt.plot(x, y, label=f'f(x) = x - 4x + 3')
plt.plot(vertex_x, vertex_y, 'ro', label=f'Vertex ({vertex_x:.1f}, {vertex_y:.1f})')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.legend()
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

3. Exponential growth of model complexity vs. performance.

```python
# Exponential function: f(x) = 2 * e^(0.5x)
def exponential_function(x, a=2, k=0.5):
    """
    Exponential function
    Growth rate determined by k
    """
    return a * np.exp(k * x)

x = np.linspace(0, 5, 100)
y = exponential_function(x)

plt.plot(x, y, label='f(x) = 2e^(0.5x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Application: Learning rate decay
epochs = np.arange(0, 100)
initial_lr = 0.1
decay_rate = 0.05
lr_schedule = initial_lr * np.exp(-decay_rate * epochs)

plt.plot(epochs, lr_schedule)
plt.title('Exponential Learning Rate Decay')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.grid(True, alpha=0.3)
plt.show()
```

4. Logarithmic function

```python
# Logarithmic function
x = np.linspace(0.01, 10, 100)
y_natural = np.log(x)  # Natural log (base e)
y_log10 = np.log10(x)  # Common log (base 10)

plt.plot(x, y_natural, label='ln(x)')
plt.plot(x, y_log10, label='log(x)')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=1, color='r', linestyle='--', alpha=0.5, label='x=1')
plt.grid(True, alpha=0.3)
plt.legend()
plt.title('Logarithmic Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

5. Calculating Slope 

```python
def calculate_slope(x1, y1, x2, y2):
    """Calculate slope between two points"""
    return (y2 - y1) / (x2 - x1)

# Example: Calculate slope of linear function
x1, y1 = 0, 3
x2, y2 = 5, 13
slope = calculate_slope(x1, y1, x2, y2)
print(f"Slope between ({x1}, {y1}) and ({x2}, {y2}): {slope}")
```

6. Evaluating limits

```python
x = sp.Symbol('x')

# Example 1: Direct substitution
f1 = x**2 + 3*x + 2
limit1 = sp.limit(f1, x, 2)
print(f"lim (x+3x+2) as x2 = {limit1}")  # Output: 12

# Example 2: Indeterminate form (0/0) - factor and simplify
f2 = (x**2 - 4) / (x - 2)
limit2 = sp.limit(f2, x, 2)
print(f"lim (x-4)/(x-2) as x2 = {limit2}")  # Output: 4

# Example 3: Limit at infinity
f3 = (3*x**2 + 2*x) / (x**2 - 1)
limit3 = sp.limit(f3, x, sp.oo)
print(f"lim (3x+2x)/(x-1) as x = {limit3}")  # Output: 3
```

7. Evaluating Continuity

```python
from typing import Callable


# Example: Checking continuity
def is_continuous_at_point(func: Callable, point, tolerance=1e-6):
    """
    Check if function is continuous at a point
    """
    try:
        # Check if function is defined at point
        f_a = func(point)
        
        # Approximate limit using nearby points
        h = 1e-7
        left_limit = func(point - h)
        right_limit = func(point + h)
        
        # Check if limit exists and equals function value
        if abs(left_limit - right_limit) < tolerance:
            limit_value = (left_limit + right_limit) / 2
            return abs(limit_value - f_a) < tolerance
        return False
    except:
        return False

# Test continuous function
f_continuous = lambda x: x**2 + 2*x + 1
print(f"f(x)=x+2x+1 continuous at x=1: {is_continuous_at_point(f_continuous, 1)}")

# Test discontinuous function
def f_discontinuous(x):
    if x != 2:
        return (x**2 - 4) / (x - 2)
    return 0  # Defined but not continuous

print(f"Discontinuous function at x=2: {is_continuous_at_point(f_discontinuous, 2)}")
```
