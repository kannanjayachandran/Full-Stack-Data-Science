```python
def bs_helper(nums: list[int], target: int, start: int, end: int) -> int:

    if nums[start] > nums[end]: return -1

    while start <= end: 
        mid = start + (end-start) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            end = mid-1
        else:
            start = mid+1
    return -1
```

**Q1 Modified binary search to find the Lower bound**

```python
def lowerBound(nums: list[int], k: int) -> int:

    n = len(nums)
    start, end, ans = 0, n-1, -1

    while start <= end:
        mid = start + (end-start) // 2
        if nums[mid] == k:
            ans = mid
            end = mid-1
        elif nums[mid] > k:
            end = mid-1
        else:
            start = mid+1
    return ans
```

```python
lowerBound([1, 1, 2, 2, 2, 2, 3, 4, 4, 5, 5, 6, 7], 2)
```

**Q3 Find the first and last occurrence of a number in a sorted array**

```python
def searchOccurrence(nums: list[int], k: int, isFirst: bool) -> int:

    start, end, ans = 0, len(nums)-1, -1

    while start <= end:
        mid  = start + (end-start) // 2
        if nums[mid] == k:
            ans = mid
            if isFirst: end = mid-1
            else: start = mid+1
        elif nums[mid] > k:
            end = mid-1
        else:
            start = mid+1
    return ans



def solution(nums: list[int], k: int) -> list[int]:

    ans = [-1, -1]
    ans[0] = searchOccurrence(nums, k, True)
    if ans[0] != - 1:
        ans[1] = searchOccurrence(nums, k , False)
    
    return ans
```

```python
solution([1, 2, 3, 4, 5, 6, 7, 7, 7, 7, 8, 9], 7)
```

**Q5 Find the number of times a number occurs in a sorted array**

```python
def countInSorted(nums: list[int], target: int) -> int:
    firstOccurrence = searchOccurrence(nums, target, True)
    if firstOccurrence != -1:
        secondOccurrence = searchOccurrence(nums, target, False)
        return (secondOccurrence - firstOccurrence) + 1
    return -1
```

```python
countInSorted([1, 2, 3, 4, 5, 6, 7, 7, 7, 7, 8, 9], 7)
```

**Q6 Find the number of times a sorted array is rotated**

```python
def sortedRotationCount(nums: list[int]) -> int:

    n = len(nums)
    start, end, ans = 0, n-1, 0
    if nums[start] <= nums[end]: return ans

    while start <= end:
        mid = start + (end-start) // 2
        nxt = (mid + 1) % n
        prev = (mid + n - 1) % n

        if nums[nxt] >= nums[mid] and nums[prev] >= nums[mid]:
            return mid
        if nums[0] <= nums[mid]:
            start = mid+1
        elif nums[n-1] >= nums[mid]:
            end = mid-1
    return -1 
```

```python
sortedRotationCount([4,5,6,7,0,1,2])
```

**Q7 Searching in sorted-rotated array** 

> If present return the index of the element, else return -1. There will be at least one element in the array and all elements are unique

```python
def search_sorted_rotated(nums: list[int], target: int) -> int:

    # array of size 1
    if len(nums) == 1:
        if nums[0] == target:
            return 0
        return -1
    
    # already sorted array
    if nums[0] <= nums[-1]:
        return bs_helper(nums, target, 0, len(nums)-1)
    
    pivotIdx = sortedRotationCount(nums)
    ans = bs_helper(nums, target, 0, pivotIdx-1)

    if ans == -1:
        ans = bs_helper(nums, target, pivotIdx, len(nums)-1)
    return ans
```

```python
search_sorted_rotated([4, 5, 6, 7, 0, 1, 2], 1)
```

**Q8 Search in nearly sorted array**

```python
def search_in_nearly_sorted(nums: list[int], target: int) -> int:
    start, end = 0, len(nums)-1

    while start <= end:
        midIdx = start + (end-start) // 2
        if nums[midIdx] == target:
            return midIdx
        if midIdx > start and nums[midIdx - 1] == target:
            return midIdx - 1
        if midIdx < end and nums[midIdx+1] == target:
            return midIdx + 1
        elif nums[midIdx] > target:
            end = midIdx-2
        else:
            start = midIdx + 2
    return -1 
```

```python
search_in_nearly_sorted([10, 3, 40, 20, 50, 80, 70], 3)
```

**Q9 Find the floor of a number in a sorted array**

```python
def find_floor(nums: list[int], target: int) -> int:
    start, end = 0, len(nums)-1

    while start <= end:
        midIdx = start + (end-start) // 2
        if nums[midIdx] > target:
            end = midIdx - 1
        elif nums[midIdx] < target:
            start = midIdx + 1
        else:
            return nums[midIdx]
    return nums[end]
```

```python
find_floor([1, 2, 3, 4, 5, 10, 20], 19)
```

```python
def find_ceil(nums: list[int], target: int) -> int:
    
    start, end = 0, len(nums)-1

    while start <= end:
        midIdx = start + (end-start) // 2
        if nums[midIdx] > target:
            end = midIdx - 1
        elif nums[midIdx] < target:
            start = midIdx + 1
        else:
            return nums[midIdx]
    return nums[start]
```

**Q11 Next alphabet**

```python
def find_next_great(letters: list[str], target: str) -> str:
    
    start, end = 0, len(letters)-1
    while start <= end:
        midIdx = start + (end-start) // 2
        if letters[midIdx] > target:
            end = midIdx -1
        else: 
            start = midIdx + 1
    return letters[start % len(letters)]
```

```python
find_next_great(['c','f','j', 'k', 'm', 'p', 'r', 'w', 'y'], 'z')
```

**Q12 Search in infinite sorted array**

> Do not use len() function or anything similar

```python
def find_search_space(nums: list[int], target: int) -> list[int]:
    start, end = 0, 1
    
    while nums[start] < target:
        start = end + 1
        end *= 2
    return [start, end]


def infinite_search(nums: list[int], target: int) -> int:

    search_range = find_search_space(nums, target)
    start = search_range[0]
    end = search_range[1]
    return bs_helper(nums, target, start, end)
```

```python
infinite_search([1, 3, 4, 6, 7, 9], 4)
```

**Q13 Find the first occurrence of 1 in a sorted binary array**

```python
def find_first_one(nums: list[int]) -> int:
    
    search_range = find_search_space(nums, 1)
    start = search_range[0]
    end = search_range[1]

    possible_ans = -1
    while start <= end:
        midIdx = start + (end-start) // 2
        if nums[midIdx] == 1:
            possible_ans = midIdx
            end = midIdx - 1
        elif nums[midIdx] > 1:
            end = midIdx - 1
        else:
            start = midIdx + 1
    return possible_ans
```

**Q14 Minimum different element**

```python
def min_diff_element(nums: list[int], key: int) -> int:

    if key < nums[0]:
        return nums[0]
    elif key > nums[-1]:
        return nums[-1]
    
    start, end = 0, len(nums)-1

    while start <= end:
        mid = start + (end-start) // 2
        if nums[mid] == key:
            return nums[mid]
        elif nums[mid] > key:
            end = mid - 1
        else:
            start = mid + 1
    
    upper_bound = abs(nums[start] - key)
    lower_bound = abs(nums[end] - key)

    return nums[end] if lower_bound <= upper_bound else nums[start]
```

```python
min_diff_element([2, 5, 10, 12, 15], 13)
```

**Q15 Peak index in a mountain array**

```python
def findPeak(nums: list[int]) -> int:
    
    start, end = 0, len(nums)-1

    while start < end:
        mid = start + (end-start) // 2
        if nums[mid] < nums[mid+1]:
            start = mid + 1
        else:
            end = mid
    return start
```

```python
findPeak([1, 2, 4, 5, 10, 7, 6, 3, 2, 1])
```

**Q16 Find the maximum element in a bitonic / mountain array**

```python
def max_in_mountain(nums: list[int]) -> int:
    return nums[findPeak(nums)]
```

**Q17 Search in Bitonic / mountain array**

```python
def bs_helper(nums: list[int], target: int, start: int, end: int) -> int:
    
    while start <= end:
        mid = start + (end - start) // 2
        if nums[mid] == target:
            return mid
        
        if nums[start] <= nums[mid]:  # Check if the left sub-array is sorted
            if nums[start] <= target < nums[mid]:
                end = mid - 1
            else:
                start = mid + 1
        else:  # The right sub-array is sorted
            if nums[mid] < target <= nums[end]:
                start = mid + 1
            else:
                end = mid - 1
    
    return -1


def search_in_mountain(nums: list[int], key: int) -> int:
   
    peak_idx = findPeak(nums)
    left_res = bs_helper(nums, key, 0, peak_idx)
    
    if left_res != -1:
        return left_res
    
    right_res = bs_helper(nums, key, peak_idx+1, len(nums)-1)
    return right_res
```

```python
search_in_mountain([1, 3, 6, 8, 12, 11, 9, 7, 4, 2], 9)
```

**Q19 Search in sorted matrix**

```python
def search_in_sorted_matrix(mat: list[list[int]], key: int) -> list[int]:

    row, col = 0, len(mat[0])-1

    while row < len(mat) and col >= 0:
        if mat[row][col] == key:
            return [row, col]
        elif mat[row][col] > key:
            col -= 1
        else:
            row += 1
    return [-1, -1]
```

```python
matrix = [
    [1, 2, 3, 4, 5],
    [6, 7, 5, 4, 3],
    [5, 6, 7, 8, 9]
]

search_in_sorted_matrix(matrix, 4)
```

**Q20 Find the minimum and maximum element in sorted-rotated array**

```python
def min_max_sorted_rotated(nums: list[int]) -> list[int]:

    min_idx = sortedRotationCount(nums)
    if min_idx != 0:
        max_idx = min_idx - 1
        return [min_idx, max_idx]
    return [min_idx, min_idx]
```
